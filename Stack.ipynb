# Stack function:

```

#include<iostream>
#include<cmath>
using namespace std;

class Stack
{
private:
    int *stack_array;
    int size;
    int capacity;
    int top;
    int last;
    bool Isempty;

    void resize()
    {
        capacity = 2*capacity;
        int *temp = new int[capacity];
        for(int i=0; i<size; i++)
        {
            temp[i] = stack_array[i];
        }
        delete []stack_array;
        stack_array = temp;
    }

public:
    Stack()
    {
        stack_array = new int[1] ;
        size = 0;
        capacity = 1;
        top = 0;
        last = 0;
        Isempty = 0;
    }

    ~Stack()
    {
        delete []stack_array;
        size = -1;
        capacity = -1;
        top = -1;
        last = -1;
        Isempty = 1;
    }

    void insert(int value)
    {
        if(size==capacity)
        {
            resize();
        }

        if(capacity < 0)
        {
          cout<<"Memory not available"<<endl;
          return;
        }

        stack_array[size] = value;
        size++;
    }

    void getstack()
    {
        for(int i=0; i<size; i++)
        {
            cout<<stack_array[i]<< " ";
            top = stack_array[0];
            last = stack_array[size-1];
        }
        cout<< endl<< "Capacity: "<< capacity <<"    "<<"Size :"<<size<<endl;
    }

    void front_element()
    {
        cout<<"top Element : ";
        cout<< top<<endl;
    }

    void back_element()
    {
        cout<<"Last Element : ";
        cout<< last<<endl;
    }

    void clear()
    {
        delete []stack_array;
        size = 0;
        capacity = 1;
        top = -1;
        last = -1;
        Isempty = 1;
    }

    void IsEmpty()
    {
        Isempty = size==0?1:0;
        cout<<Isempty<<endl;
    }

    void pop()
    {
        for(int i=0; i<size; i++)
        {
            stack_array[i] = stack_array[i+1];
        }
        stack_array[size-1] = -1;
        size--;
        if(size <= capacity/2)
        {
            capacity = capacity/2;
        }

    }

    void search(int value)
    {
        for(int i=0; i<size; i++)
        {
            if(stack_array[i] == value)
            {
                cout<<"value searching for "<<value<< " exist"<<endl;
                return;
            }
        }
        cout<<"value does not exist"<<endl;
    }

};
int main()
{
    Stack s;
    int choice, value;
    cout<<" List : "<<endl;
    cout<<"-------------------------------------------"<<endl;
    cout<<"Enter 1 to push/insert value into the array"<<endl;
    cout<<"Enter 2 to display the array"<<endl;
    cout<<"Enter 3 to display first element of the array"<<endl;
    cout<<"Enter 4 to display last element of the array"<<endl;
    cout<<"Enter 5 to pop the front element array "<<endl;
    cout<<"Enter 6 to check array is empty(1) or not empty(0)"<<endl;
    cout<<"Enter 7 to clear the entire array "<<endl;
    cout<<"Enter 8 to search the element in array "<<endl;


    cout<<endl;

    while(1)
    {
       cout <<endl<<"Enter your choice: ";
       cin >> choice;
       switch(choice)
        {
            cout<<endl;
            case 1:
                cout << "Enter value to push: ";
                cin >> value;
                s.insert(value);
                break;

            case 2:
                cout << "Stack Display: ";
                s.getstack();
                break;

            case 3:
                s.front_element();
                break;

            case 4:
                s.back_element();
                break;

            case 5:
                s.pop();
                cout << "Pop successful! ";
                break;

            case 6:
                s.IsEmpty();
                break;

            case 7:
                s.clear();
                cout << "Stack Cleared!: ";
                break;

            case 8:
                cout << "Enter value to seach: ";
                cin >> value;
                s.search(value);
                break;


             default:
                cout << "Invalid entry, please try again --> program exit." << endl;
                return 0;
        }
    }
    return 0;
}

```
