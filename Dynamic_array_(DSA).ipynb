# Dynamic Array Function:

```

#include<iostream>
using namespace std;

class dynamicArray
{
private:
    int Capacity;
    int Size;
    int *arr;
    int top;
    int last;
    bool IsEmpty;

public:
    dynamicArray()
    {
        Size = 0;
        Capacity = 1;
        arr = new int[1];
        top = 0;
        last = 0;
        IsEmpty = 1;
    }

    ~dynamicArray()
    {
        Size = 0;
        Capacity = 0;
        delete []arr;
        top = 0;
        last = 0;
        IsEmpty = 1;
    }

    void resize()
    {
        Capacity = 2*Capacity;
        int *temp = new int[Capacity];
        for(int i = 0; i<Size; i++)
        {
            temp[i] = arr[i];
        }
        delete []arr;
        arr = temp;
    }

    void push(int value)
    {
        if(Size == Capacity)
        {
            resize();
        }
        if(Capacity < 0)
        {
            cout<<"Out of memory"<<endl;
            return;
        }
        arr[Size] = value;
        Size++;
    }

    void getarry()
    {
        cout<<"Array :"<< "";
        for(int i=0; i<Size; i++)
        {
            cout<<arr[i]<< " ";
            top = arr[0];
            last = arr[Size-1];
        }
        cout<< endl<< "Capacity: "<< Capacity <<"    "<<"Size :"<<Size<<endl;
    }

    void top_element()
    {
        cout<<"top Element : ";
        cout<< top<<endl;
    }

    void back_element()
    {
        cout<<"Last Element : ";
        cout<< last<<endl;
    }

    void insert_at(int index, int value)
    {
        int temp = value;
        if(Capacity < 0 || Size < index || Size <= 0)
        {
            cout<<"Index does not exit"<<endl;
            return;
        }
        if(Size == Capacity)
        {
            resize();
        }
        Size++;
        index--;
        for(int i = Size-1; i>=index; i--)
        {
            if(i >= index)
            {
                arr[i+1] = arr[i];
            }
        }
        arr[index] = value;
    }

    void pop_index(int index)
    {
        index--;
        for(int i=0; i<Size; i++)
        {
            if(i >= index)
            {
                arr[i] = arr[i+1];
            }
        }
        arr[last]= -1;
        Size--;
        if(Size <= Capacity / 2)
        {
            Capacity = Capacity/2;
        }

    }
    void search(int value)
    {
        for(int i=0; i<Size; i++)
        {
            if(arr[i] == value)
            {
                cout<<"value searching for "<<value<< " exist"<<endl;
                return;
            }
        }
        cout<<"value does not exist"<<endl;
    }

    void Isempty()
    {
        // empty is 1 and not empty is 0
        IsEmpty = Size==0?1:0;
        cout<<IsEmpty<<endl;
    }

    int clear()
    {
        //this->~dynamicArray(); //dangling pointer never use this
        delete[] arr;
        arr = nullptr;
        Size = 0;
        Capacity = 1;
        IsEmpty = true;
    }
};

int main()
{
    dynamicArray a;
    int choice, value, index;
    cout<<" List : "<<endl;
    cout<<"___________________________________________________"<<endl;
    cout<<"Enter 1 to push/insert value into the array"<<endl;
    cout<<"Enter 2 to pop value from the array"<<endl;
    cout<<"Enter 3 to clear the entire array"<<endl;
    cout<<"Enter 4 to display the array"<<endl;
    cout<<"Enter 5 to insert at specific index into the array"<<endl;
    cout<<"Enter 6 to display first element of the array"<<endl;
    cout<<"Enter 7 to display last element of the array"<<endl;
    cout<<"Enter 8 to check array is empty(1) or not empty(0)"<<endl;
    cout<<"Enter any other key to exit the program "<<endl;
    cout<<endl;
    cout<<"Select from list : "<<endl;
    cout<<endl;

    while(1)
    {
       cout <<endl<<"Enter your choice: ";
       cin >> choice;

       switch(choice)
        {
            case 1:
                cout << "Enter value to push: ";
                cin >> value;
                a.push(value);
                break;

            case 2:
                cout << "Enter index to pop: ";
                cin >> index;
                a.pop_index(index);
                break;

            case 3:
                a.clear();
                cout << "Array cleared!" << endl;
                break;

            case 4:
                a.getarry();
                break;

            case 5:
                cout << "Enter index to insert at: ";
                cin >> index;
                cout << "Enter value to insert: ";
                cin >> value;
                a.insert_at(index, value);
                break;

            case 6:
                a.top_element();
                break;

            case 7:
                a.back_element();
                break;

            case 8:
                a.Isempty();
                break;

            default:
                cout << "Invalid entry, please try again --> program exit." << endl;
                return 0;
        }
    }

    return 0;
}


```
