# Binary Tree Menu : -----------------------------------------------------------


```

#include<iostream>
using namespace std;

class Node
{
public:
    int value;
    Node *left;
    Node *right;

    Node(int value) : value(value), right(nullptr), left(nullptr) {};

    ~Node()
    {
        delete left;
        delete right;
    }

    Node* Minimum_Node()
    {
        Node *temp = this;
        while (temp != nullptr && temp->left != nullptr)
        {
            temp = temp->left;
        }
        if (temp) cout << "Minimum node value: " << temp->value << endl;
        return temp;
    }

    Node* Maximum_Node()
    {
        Node *temp = this;
        while (temp != nullptr && temp->right != nullptr)
        {
            temp = temp->right;
        }
        if (temp) cout << "Maximum node value: " << temp->value << endl;
        return temp;
    }

    Node* Insert(Node *root, int insert_value)
    {
        if (root == nullptr)
        {
            return new Node(insert_value);
        }

        if (root->value > insert_value) // go left
        {
            root->left = Insert(root->left, insert_value);
        }
        else if (root->value < insert_value) // go right
        {
            root->right = Insert(root->right, insert_value);
        }

        return root;
    }

    Node* Delete(Node* root, int value)
{
    if (root == nullptr) return nullptr;

    if (value < root->value)
        root->left = Delete(root->left, value);
    else if (value > root->value)
        root->right = Delete(root->right, value);
    else
    {
        // Case 1: No children
        if (root->left == nullptr && root->right == nullptr)
        {
            delete root;
            return nullptr;
        }
        // Case 2: One child (right child only)
        else if (root->left == nullptr)
        {
            Node* temp = root->right;
            delete root;
            return temp;
        }
        // Case 3: One child (left child only)
        else if (root->right == nullptr)
        {
            Node* temp = root->left;
            delete root;
            return temp;
        }
        // Case 4: Two children
        else
        {
            Node* temp = root->right;
            while (temp->left != nullptr)
            {
                temp = temp->left;
            }
            root->value = temp->value;
            root->right = Delete(root->right, temp->value);
        }
    }

    return root;
}


    void Inorder()
    {
        if(this !=nullptr)
        {
            if (left)left->Inorder();
            cout<<this->value<<" ";
            if (right)right->Inorder();
        }
    }
};

int main()
{
    Node *root = nullptr;
    int choice, value;
    cout << "--------------------Menu:----------------------------------\n";
    cout << "1. Insert Node\n";
    cout << "2. Find Minimum Node\n";
    cout << "3. Find Maximum Node\n";
    cout << "4. Exit Code\n";
    cout << "5. InOrder printing \n";
    cout << "6. Delete Value\n";


    while(true)
    {
        cout<<endl;
        cout << "Enter your choice: ";
        cin >> choice;
        switch (choice)
        {
        case 1:
            cout << "Enter value to insert: ";
            cin >> value;
            root = root->Insert(root, value);
            break;


        case 2:
            if (root)
                root->Minimum_Node();
            else
                cout << "Tree is empty." << endl;
            break;

        case 3:
            if (root)
                root->Maximum_Node();
            else
                cout << "Tree is empty." << endl;
            break;

        case 4:
            cout << "Exiting..." << endl;
            return 0;

        case 5:
            cout<<"\nInOrder Printing: (LNR) ";
            root->Inorder();
            cout<<endl;
            break;

        case 6:
            cout << "Delete value: ";
            cin >> value;
            root = root->Delete(root, value);
            break;

        default:
            cout << "Invalid choice. Please try again." << endl;
        }
    }
    delete root;
    return 0;
}

```
