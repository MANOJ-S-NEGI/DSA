======================== **Double Ended Queue** ================================


```

#include <iostream>
using namespace std;

class DeQue
{
public:
    int capacity;
    int *arr;
    int front_idx;
    int rear_idx;
    int size;

    // Constructor
    DeQue(int cap = 20) : capacity(cap), front_idx(0), rear_idx(cap), size(0)
    {
        arr = new int[capacity];
    }

    ~DeQue()
    {
        delete[] arr;
    }

    void enqueue_front()
    {
        if (size == capacity)
        {
            cout << "Queue is overflow!" << endl;
            return;
        }
        cout << "Insert into front: ";
        cin >> arr[front_idx];
        front_idx++;
        size++;
    }

    void enqueue_rear()
    {
        if (size == capacity)
        {
            cout << "Queue is overflow!" << endl;
            return;
        }
        cout << "Insert into rear: ";
        cin >> arr[rear_idx-1];
        rear_idx--;
        size++;
    }

    void getDeque() const
    {
        if (size == 0)
        {
            cout << "Queue is empty!" << endl;
            return;
        }

        cout << "Queue elements: ";
        for (int i = 0 ; i<front_idx; i++)
        {
            cout << arr[i] << " ";
        }
        cout<< "| ";
        for (int i = rear_idx ; i<capacity; i++)
        {
            cout << arr[i] << " ";
        }
        cout << endl << "Size: " << size << "  Capacity: " << capacity << endl;
        //cout << "FI: " << front_idx << "  rearIndex: " << rear_idx << endl;
    }


    void dequeue_front()
    {
        if (size == 0)
        {
            cout << "Queue is underflow!" << endl;
            return;
        }

        if(front_idx==0)
        {
            int del_val = arr[rear_idx];
            arr[rear_idx] = -1;
            rear_idx++;
            size--;
            cout<<"Element value: "<<del_val<<" deleted"<<endl;
            return;
        }

        int del_val = arr[0];
        for (int i = 0 ; i<front_idx; i++)
        {
            arr[i] = arr[i+1];
        }
        cout<<"Element value: "<<del_val<<" deleted"<<endl;
        arr[front_idx+1]== -1;
        front_idx--;
        size--;

    }

    void dequeue_rear()
    {
        if (size == 0)
        {
            cout << "Queue is underflow!" << endl;
            return;
        }

        if(rear_idx==20)
        {
            int del_val = arr[front_idx-1];
            arr[front_idx-1] = -1;
            front_idx--;
            size--;
            cout<<"Element value: "<< del_val <<" deleted"<<endl;
            return;
        }

        int del_val = arr[capacity-1];
        for (int i = capacity-1 ; i>rear_idx; i--)
        {
            arr[i] = arr[i-1];
        }
        cout<<"Element value: "<< del_val <<" deleted"<<endl;
        arr[rear_idx]=-1;
        rear_idx++;
        size--;
    }
};

int main()
{
    cout << "Choose from List: " << endl;
    cout << "===============================================" << endl;
    cout << "Enter 1 -> To insert the element from front" << endl;
    cout << "Enter 2 -> To insert the element from rear" << endl;
    cout << "Enter 3 -> To display the queue" << endl;
    cout << "Enter 4 -> To delete front of the queue" << endl;
    cout << "Enter 5 -> To delete rear of the queue" << endl;
    cout << "Enter 6 -> To exit program" << endl;

    cout << "===============================================" << endl;

    int choice;
    DeQue q;

    while (true)
    {
        cout << endl
             << "Enter Action: ";
        cin >> choice;
        switch (choice)
        {
            case 1:
                q.enqueue_front();
                break;
            case 2:
                q.enqueue_rear();
                break;
            case 3:
                q.getDeque();
                break;
            case 4:
                q.dequeue_front();
                break;
            case 5:
                q.dequeue_rear();
                break;

            case 6:
                cout << "Exiting program.! " << endl;
                return 0;

            default:
                cout << "Invalid choice ! Try again " << endl;
                break;

        }
    }

    return 0;
}


```

======================== **Single Ended Queue** ================================


```

#include<iostream>
using namespace std;

class Queue
{
public:
    int capacity;
    int size;
    int *arr ;
    int rear;
    int front;

    Queue():capacity(1), size(0), rear(0), front(0){ arr = new int[capacity];}

    void enqueue()
    {
        if(rear == capacity)
        {
            capacity = 2*capacity;
            int *temp = new int[capacity];
            for(int i = 0; i<rear; i++)
            {
                temp[i] = arr[i];
            }
            rear -= front;
            front = 0;
            delete []arr;
            arr = temp;
        }
        cout<<"Insert into queue: ";
        cin>>arr[rear];
        rear++;
        size++;
    }
    void dequeue()
    {
        if(size == 0)
        {
            cout<< "Queue is empty !";
            front = 0;
            rear = 0;
            return;
        }
        int del_val = arr[front];
        front++;
        size--;
        cout<< "element "<<del_val<<" deleted"<< endl;
        //cout<<"front ->"<<front<<endl;
        //cout<<"rar ->"<<rear<<endl;

        if(size==capacity/4)
        {
            capacity = capacity/4;
            int *tempop = new int[capacity];
            for(int i = 0; i<size; i++)
            {
                tempop[i] = arr[front];
                front++;
            }
            delete []arr;
            front = 0;
            rear = size;
            arr = tempop;
        }
    }

    void getQue()
    {
        if (size == 0)
        {
            cout << "Queue is empty!" << endl;
            return;
        }
        cout << "Queue elements: ";
        for(int i = front; i<size; i++)
        {
            cout<< arr[i]<<" ";
        }
        cout<<endl<< "size: "<<size<< "     capacity: "<<capacity<<endl;
    }
};

int main()
{
    cout << "Choose from List: " << endl;
    cout << "===============================================" << endl;
    cout << "Enter 1 -> To insert the element" << endl;
    cout << "Enter 2 -> To display the queue" << endl;
    cout << "Enter 3 -> To delete front of the queue" << endl;
    cout << "===============================================" << endl;

    int choice;
    Queue q;

    while (true)
    {
        cout << endl << "Enter Action: ";
        cin >> choice;
        switch (choice)
        {
            case 1:
                q.enqueue();
                break;

            case 2:
                q.getQue();
                break;

            case 3:
                q.dequeue();
                break;

            default:
                cout << "Exiting program." << endl;
                return 0;
        }
    }

    return 0;
}


```
