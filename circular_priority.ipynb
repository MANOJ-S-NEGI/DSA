============================= Circular Priority Queue ==========================

```

#include<iostream>
using namespace std;

class PriorityCircular
{
public:
    int rear;
    int front;
    int capacity;
    int size;
    int *arr;

    PriorityCircular():rear(-1), front(-1), capacity(10), size(0){arr = new int[capacity];}
    ~PriorityCircular()
    {
         delete[] arr;
    }

    void insert_(int val)
    {
        if(size == capacity)
        {
            cout << "Queue is full!" << endl;
            return;
        }

        if(front == -1 && rear == -1)
        {
            front = 0;
            rear = 0;
            arr[rear] = val;
        }
        else
        {
            arr[rear]= val;
            for(int i = 0; i<=size ;i++)
            {
                if(arr[rear] > arr[i])//dry run : [6,5,4,2,1] val= 5
                {
                   int temp = arr[i];
                   arr[i] = arr[rear];
                   arr[rear] = temp;
                }
            }
        }
        rear = (rear+1)%capacity;
        size++;
    }

    void pop()
    {
        if (size == 0)
        {
            cout << "Queue is underflow!" << endl;
            return;
        }
        cout<< "element "<<arr[front]<<" deleted"<<endl;
        arr[front] = -1;
        front = (front+1)%capacity;
        size--;

        if (size == 0)
        {
            front = -1;
            rear = -1;
        }
    }

    void getQue()
    {
        if (size == 0)
        {
            cout << "Queue is empty!" << endl;
            return;
        }

        cout << "Queue elements: ";
        for (int i = 0; i < size; i++)
        {
            cout << arr[(front + i) % capacity] << " ";
        }
        cout << endl;
        cout << "Size: " << size << " Capacity: " << capacity << endl;
        //cout << "Rear: " << rear << ", Front: " << front << endl;
    }
};

int main()
{
    cout << "======This is Circular priority Queue ======== " << endl;
    cout << "===============================================" << endl;
    cout << "Choose from List: " << endl;
    cout << "===============================================" << endl;
    cout << "Enter 1 -> To insert the element" << endl;
    cout << "Enter 2 -> To display the queue" << endl;
    cout << "Enter 3 -> To delete element of queue" << endl;
    cout << "===============================================" << endl;

    PriorityCircular pc;
    int choice;
    while(true)
    {
        cout << endl
             << "Enter Action: ";
        cin >> choice;
        switch(choice)
        {
            case 1:
                int val;
                cout<<"Enter into queue : ";
                cin>> val;
                pc.insert_(val);
                break;

            case 2:
                pc.getQue();
                break;

            case 3:
                pc.pop();
                break;

            default:
                cout << "Invalid choice ! Try again " << endl;
                break;
        }
    }
    return 0;
}


```
